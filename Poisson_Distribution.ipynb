```python
data_binom = poisson.rvs(mu=4.75, size=44685)
ax = sb.distplot(data_binom,
                  kde=True,
                  color='red',
                  hist_kws={"linewidth": 25,'alpha':1})
ax.set(xlabel='Poisson', ylabel='Frequency', title='Poisson Distribution, Probability of a collision in a month on the N1, given an average of 44,685 vehicles, and an average of 4.75 collisions')

#Poisson Distribution, using the average collisions monthly and average vehicles monthly on the N1
```

```python
data_binom = poisson.rvs(mu=4.0375, size=44685)
ax = sb.distplot(data_binom,
                  kde=True,
                  color='blue',
                  hist_kws={"linewidth": 25,'alpha':1})
ax.set(xlabel='Poisson', ylabel='Frequency', title='Poisson Distribution, Probability of a collision in a month on the N1, given an average of 44,685 vehicles, and an average of 4.0375 collisions')

#Poisson Distribution, using the average collisions monthly and average vehicles monthly on the N1
#Collisions reduced by 10% to mirror effect of variable speed limits on national roads
```
